```python
import pandas as pd
import io
import requests

# Актуальная ссылка на экспорт данных о поездках (Dockless Vehicles)
# Мы берем последние данные через фильтр Socrata
URL = "https://data.louisvilleky.gov/api/views/v7r2-iswv/rows.csv?accessType=DOWNLOAD"
OUTPUT_FILE = "louisville_trips.parquet"

def download_louisville_data(limit=10000):
    print(f"Запрашиваю данные напрямую из Луисвилла...")
    
    try:
        # Добавляем лимит строк через параметры запроса SODA
        # и заголовки, чтобы сервер не принял нас за бота
        params = {'$limit': limit}
        headers = {'User-Agent': 'Mozilla/5.0'}
        
        response = requests.get(URL, params=params, headers=headers)
        response.raise_for_status() # Проверка на 200 OK
        
        # Читаем CSV из потока байтов
        df = pd.read_csv(io.BytesIO(response.content))
        
        if df.empty:
            print("Сервер вернул пустой файл.")
            return None

        # Приводим названия колонок к нижнему регистру для удобства
        df.columns = [c.lower() for c in df.columns]

        # Конвертация типов (важно для ГИС)
        cols_to_fix = ['startlatitude', 'startlongitude', 'endlatitude', 'endlongitude', 'tripdistance']
        for col in cols_to_fix:
            if col in df.columns:
                df[col] = pd.to_numeric(df[col], errors='coerce')

        # Сохраняем в компактный parquet
        df.to_parquet(OUTPUT_FILE, index=False)
        
        print(f"Успешно! Загружено строк: {len(df)}")
        print(f"Файл сохранен: {OUTPUT_FILE}")
        return df

    except Exception as e:
        print(f"Произошла ошибка: {e}")
        return None

if __name__ == "__main__":
    download_louisville_data(limit=20000)
```
